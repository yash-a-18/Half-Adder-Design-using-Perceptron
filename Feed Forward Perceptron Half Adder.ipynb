# Final Goal: Half Adder circuit

**Importing the required libraries**

```python
import numpy as np
```

**Unit Step function - Activation Function**

```python
def unitStep(v):
    try:
        if len(v) > 1:
            if v.all() >= 0:
                return 1
            else:
                return 0
    except:
        if v >= 0:
            return 1
        else:
            return 0
```

**Perceptron Model**

```python
def perceptronModel(x, w, b):
    v = np.dot(w, x) + b
    y = unitStep(v)
    return y
```

**NOT Gate**

```python
def NOT_GATE(x):
    wNOT = -1
    bNOT = 0.5
    return perceptronModel(x, wNOT, bNOT)
```

**AND Gate**

```python
def AND_GATE(x):
    w = np.array([1,1])
    bAND = -1.5
    return perceptronModel(x, w, bAND)
```

## Half Adder

```python
def Half_Adder(x):
    # First NAND
    # op_AND1 = AND_GATE(x)
    y1 = NOT_GATE(AND_GATE(x))
    
    carry = NOT_GATE(AND_GATE([y1,y1]))
    
    # Upper NAND
    # upper_op_AND2 = AND_GATE([x[0], y1])
    y2 = NOT_GATE(AND_GATE([x[0], y1]))
    
    # Lower NAND
    # lower_op_AND3 = AND_GATE([y1, x[1]])
    y3 = NOT_GATE(AND_GATE([y1, x[1]]))
    
    # final NAND
    # final_op_AND3 = AND_GATE([y2, y3])
    Sum = NOT_GATE(AND_GATE([y2, y3]))
    
    return Sum, carry
```

```python
test1 = np.array([0,0])
test2 = np.array([0,1])
test3 = np.array([1,0])
test4 = np.array([1,1])
test_cases = [test1, test2, test3, test4]
```

**Output**

```python
for test in test_cases:
    ans = Half_Adder(test)
    print("Half-Adder({},{}) = Sum: {} Carry: {}".format(test[0], test[1], ans[0], ans[1]))
```
